DATA CLEANING

```python
#Instead of dropping missing values from no_of_ratings & ratings ,i will be replacing them with 0
df['ratings']=df['ratings'].fillna(0)
df['no_of_ratings']=df['no_of_ratings'].fillna(0)
df.isna().sum()
```

```python
#Drop (without the drop method) missing values from actual_price and replace missing values from discount_price with 0
df=df[df['actual_price'].notna()]
df['discount_price']=df['discount_price'].fillna(0)
df.isna().sum()
```

```python
#Convert columns data types to the appropriate types
'''
 Convert ratings & no_of_ratings to Float
 1 Step convert ratings and no_of_ratings from object to string
 2 Step Assign 0 to all non digit values from ratings and no_of_ratings
 3 Step convert ratings and no_of_ratings from string to float
'''

# ratings column
df['ratings'] =df['ratings'].astype('string')
df['ratings']=df['ratings'].replace('[^0-9,.]','0',regex=True)

# no_of_ratings column ( Because no_of_ratings has ',' instead of '.' so i cannot cast it to float type and probably has more than 1 ',' i had to replace ',' with '.' and i only took 2 digits after the '.' Means 2,255 became 2.25  )
df['no_of_ratings'] =df['no_of_ratings'].astype('string')
df['no_of_ratings']=df['no_of_ratings'].replace('[^0-9,.]','0',regex=True)
df['no_of_ratings']=df['no_of_ratings'].apply(lambda x:x.split(',')[0]+'.'+x.split(',')[1][0:2] if(x.find(',')!=-1) else x)

df['no_of_ratings']=df['no_of_ratings'].astype(float)
df['ratings']=df['ratings'].astype(float)

'''
Another method would have been df['ratings'].unique()
After that delete all non digit characters ('get','FREE' etc)
'''
```

```python
'''
 Convert discount_price & actual_price to Float
 1 Check if all prices are in Rupee
 2 Remove currency sign '₹'
 3 Convert prices from string to float
 4 Because i will be doing the same steps for discount_price and actual_price , i will use a function instead of repeating code
'''

#Function 
def price_convert(column):
    df[column]=df[column].astype('string')
    df[column] =df[column].apply(lambda x: x[1:] if x[0]=='₹' else x )
    df.loc[(df[column].str[0]=='₹')]
    df[column]=df[column].replace('[^0-9.,]','0',regex=True)
    df[column]=df[column].apply(lambda x:x.split(',')[0]+'.'+x.split(',')[1][0:2] if(x.find(',')!=-1) else x)
    df[column]= df[column].astype(float)

price_convert('discount_price')
price_convert('actual_price')
```
